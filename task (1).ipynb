lottery

```python
import random
import pandas as pd

computer_wins = 0
user_wins = 0

for i in range(10):
    computer = random.randint(1, 9)
    user_input = int(input("Enter the number between 1 and 9: "))
    print(f"Computer: {computer}")
    print(f"User: {user_input}")
    if computer > user_input:
        print("Computer won")
        computer_wins += 1
    elif computer == user_input:
        print("Tie")
    else:
        print("You won")
        user_wins += 1

data = {"Computer Wins": [computer_wins], "User Wins": [user_wins]}
df = pd.DataFrame(data)

csv_filename = 'output.csv'
df.to_csv(csv_filename, index=False)

print(f"Data saved to {csv_filename}")


```

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_filename = 'output.csv'
df = pd.read_csv(csv_filename)

# Extract the win counts from the DataFrame
computer_wins = df['Computer Wins'][0]
user_wins = df['User Wins'][0]

# Define labels and values for the bar plot
labels = ['Computer Wins', 'User Wins']
values = [computer_wins, user_wins]

# Create a bar plot
plt.bar(labels, values, color=['blue', 'green'])
plt.xlabel('Outcome')
plt.ylabel('Number of Wins')
plt.title('Computer vs User Wins')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/task_1.csv')

mean_runs = data['runs'].mean()
mean_batting_avg = data['batting average'].mean()
mean_strike_rate = data['strike rate'].mean()

print(f"Mean Runs: {mean_runs}")
print(f"Mean Batting Average: {mean_batting_avg}")
print(f"Mean Strike Rate: {mean_strike_rate}")

metrics = ['Runs', 'Batting Average', 'Strike Rate']
mean_values = [mean_runs, mean_batting_avg, mean_strike_rate]

plt.plot(metrics, mean_values, marker='o')
plt.ylabel('Mean Value')
plt.title('Mean Statistics')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/task_1.csv')
numeric_columns = ["runs", "batting average", "strike rate"]
median_values = data[numeric_columns].median()

print(data.to_string())
print("\nMedian values:")
print(median_values)

median_values.plot(marker='o')
plt.ylabel('Median Value')
plt.title('Median Statistics')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/task_1.csv')
numeric_columns = ["runs", "batting average", "strike rate"]
mode_values = data[numeric_columns].mode().iloc[0]

print(data.to_string())
print("Mode values:")
print(mode_values)

mode_values.plot(marker='o')
plt.ylabel('Mode Value')
plt.title('Mode Statistics')
plt.show()

```
