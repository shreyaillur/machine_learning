```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate random data
#np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])

# Transform the features to higher degrees
poly_degree = 3
poly_features = PolynomialFeatures(degree=poly_degree)
X_poly = poly_features.fit_transform(X)

# Create and fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predict the values using the model
X_test = np.linspace(0, 5, 100).reshape(-1, 1)
X_test_poly = poly_features.transform(X_test)
y_pred = model.predict(X_test_poly)

# Calculate the mean squared error
mse = mean_squared_error(y, model.predict(X_poly))
print("Mean Squared Error:", mse)

# Plot the original data and the polynomial regression curve
plt.scatter(X, y, s=20, label="Data")
plt.plot(X_test, y_pred, color='r', label="Polynomial Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D

# Sample data: X = features, y = target
X = np.array([[1, 2, 3],
              [2, 3, 4],
              [3, 4, 5],
              [4, 5, 6],
              [5, 6, 7]])

y = np.array([5, 6, 7, 8, 9])

# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict using the model
new_X = np.array([[6, 7, 8],
                  [7, 8, 9]])
predictions = model.predict(new_X)

# Plotting the original data points in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], y, c='r', marker='o', label='Data Points')

# Plotting the regression hyperplane
x_range = np.linspace(0, 8, 100)
y_range = np.linspace(0, 9, 100)
x_range, y_range = np.meshgrid(x_range, y_range)
z_range = model.coef_[0] * x_range + model.coef_[1] * y_range + model.intercept_
ax.plot_surface(x_range, y_range, z_range, alpha=0.5, cmap='viridis', label='Regression Hyperplane')

# Plotting the predicted points
ax.scatter(new_X[:, 0], new_X[:, 1], predictions, c='g', marker='x', s=100, label='Predicted Points')

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')
ax.set_title('Multivariate Linear Regression')
plt.show()



```
