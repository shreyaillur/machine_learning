```python
def add(*bunch):
  sum=0
  for i in bunch:
    sum+=i
  return sum
print(add(1,2,3,4,5))
```

```python
def greet(**kwargs):
  print("hello")
  for key,value in kwargs.items():
    print(value)
greet(name="shreya",usn="01fe21bec030",branch="ece")
```

```python
def greet(**kwargs):
  print("hello")
  for key,value in kwargs.items():
    print(value,end="  ")
greet(name="shreya",usn="01fe21bec030",branch="ece")
```

lottery game

```python
import random
computer=random.randint(1,9)
userinput=int(input("enter the number between 1 and 9"))
print(computer)
print(userinput)
if computer>userinput:
  print("computer won")
elif computer==userinput:
  print("tie")
else:
  print("you won")
```

```python
import random
gamelist=['rock','paper','scissor']
comp=random.choice(gamelist)
user= input("enter either rock or paper or scissor")
print(comp)
print(user)
if(comp=="rock" and user=="paper"):
  print("computer won")
elif(comp=="paper" and user=="scissor"):
  print("you won")
elif(comp=="scissor" and user=="paper"):
  print("computer won")
else:
  print("tie")
```
